## Case_1

```python
#Case 1 : Function with no parameter and no return value
def factorial():
    num = int(input("Enter a number : "))
    f = 1
    while num>=1:
        f = f*num
        num-=1
    print("Factorial =",f)

factorial()
```

## Case_2

```python
#Case 2 : Function with parameter and no return value
def factorial(num):
    f = 1

    while num >= 1:
        f = f * num
        num -= 1

    print("Factorial =", f)

num = int(input("Enter a number: "))
factorial(num)
```

## Case_3

```python
#Case 3 : Function with no parameter and return value
def factorial():    
    num = int(input("Enter a number : "))
    f = 1
    while num>=1:
        f = f*num
        num-=1
    return f


result = factorial()
print("Factorial =",result)
```

## Case_4

```python
#Case 4 : Function with parameter and return value
def factorial(num):        
    f = 1
    while num>=1:
        f = f*num
        num-=1
    return f

num = int(input("Enter a number : "))
result = factorial(num)
print("Factorial =",result)
```

## Demo1

```python
#Default argument concept
#It starts from left to right
#right to left not allowed
def sum(x=0,y=0,z=0):
    return x+y+z

x= int(input("Enter a number : "))
y= int(input("Enter a number : "))
z = int(input("Enter a number : "))
s = sum(x,y,z)
print("Sum =",s)
s = sum(x,y)
print("Sum =",s)
s = sum(x)
print("Sum =",s)
s = sum()
print("Sum =",s)
```

## Keyword Argument

```python
#Keyword Argument

def display(**kwargs):
    print(kwargs)


display(eid=101,ename="Akshay",salary=32000)
display(sid=101,sname="Rahul",percentage=78,age=12)
display(bid=101,bname="Let us C",price=600,author='Kanetkar')
```
