```python
import numpy as np
a = np.array([20,30,70,49])
print(np.sum(a))
print(np.max(a))
print(np.min(a))
print(np.size(a))
print(np.size(a))
print(np.cumsum(a))
```

```python
import numpy as np

a = [100, 130, 200, 150, 200]
b = [20, 12, 50, 20, 60]

price = np.array(a)
quantity = np.array(b)

print("Price array:", price)
print("Quantity array:", quantity)
print("Sum of prices:", np.sum(price))
print()
print(np.cumsum([price,quantity]))
print()
c = np.cumprod([price,quantity], axis = 0)
print(c[1].sum())
```

```python
import numpy as np

a = [100, 130, 200, 150, 200]
b = [20, 12, 50, 20, 60]

price = np.array(a)
quantity = np.array(b)

print("Price array:", price)
print("Quantity array:", quantity)

# Sum of prices
print("Sum of prices:", np.sum(price))

# Cumulative sum of each array separately
print("Cumulative sum of prices:", np.cumsum(price))
print("Cumulative sum of quantities:", np.cumsum(quantity))

# Element-wise product of price and quantity
print("Element-wise product:", price * quantity)

# Cumulative product of each array separately
print("Cumulative product of prices:", np.cumprod(price))
print("Cumulative product of quantities:", np.cumprod(quantity))
```

```python
import numpy as np

# 3D array: shape (2, 2, 4)
# → 2 "blocks", each block has 2 rows × 4 columns
var = np.array([
    [[2, 6, 5, 3],
     [6, 5, 9, 5]],
    
    [[1, 2, 3, 4],
     [7, 8, 9, 0]]
])

print("3D Array:\n", var)
print("\nShape:", var.shape)
print()

# Loop through the 3D array
for block in var:                # 1st dimension
    print("Block:\n", block)
    print()
    for row in block:            # 2nd dimension
        print("Row:", row)
        for elem in row:         # 3rd dimension
            print("Element:", elem)
        print()
```

## Arithmetic Operation in matrix

# Add

```python
import numpy as np
var = np.matrix([[1,2,3],[1,2,3]])
var1 = np.matrix([[1,2,3],[1,2,3]])
print(var)
print()
print(var + var1)
```

# Sub

```python
import numpy as np
var = np.matrix([[1,2,3],[1,2,3]])
var1 = np.matrix([[1,2,3],[1,2,3]])
print(var)
print()
print(var - var1)
```

## Multiply

# Matrix Function

## Transpose

```python
import numpy as np
var1 = np.matrix([[1,2,3],[1,2,3]])
print(var1)
print()
print(np.transpose(var1))
```

```python
import numpy as np
var1 = np.matrix([[1,2,3],[1,2,3]])
print(var1)
print()
print(var1.T)
```

## Swapaxes

```python
import numpy as np
var1 = np.matrix([[1,2,3],[1,2,3]])
print(var1)
print()
print(np.swapaxes(var1,0,1))
```

```python
import numpy as np
var1 = np.matrix([[1,2,3],[5,4,3]])
print(var1)
print()
print(np.swapaxes(var1,0,1))
print()
```

## Inverse

```python
import numpy as np

var1 = np.matrix([[1, 2], [3, 4]])
print(var1)
print()
print(np.linalg.inv(var1))
```

## Power

```python
import numpy as np

var1 = np.matrix([[1, 2], [3, 4]])
print(var1)
print()
print(np.linalg.matrix_power(var1,3))
print()
print(np.linalg.matrix_power(var1,0))
```

## Determinate
