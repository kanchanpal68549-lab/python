```python
"""
Parking Lot Management System (Simple Version)
"""

from datetime import datetime

# ======================== Config ========================
HOURLY_RATE = 200.0
DATE_FORMAT = "%Y-%m-%d %H:%M:%S"


# ======================== Main Functions ========================
sessions = []          # List to store all parking sessions
session_id_counter = 1


def get_current_time():
    """Get current time as formatted string."""
    return datetime.now().strftime(DATE_FORMAT)


def parse_time(time_str):
    """Convert string to datetime object."""
    return datetime.strptime(time_str, DATE_FORMAT)


def add_entry():
    """Add new vehicle entry."""
    global session_id_counter
    
    vehicle_no = input("Vehicle number: ").strip()
    owner_name = input("Owner name: ").strip()
    
    if not vehicle_no or not owner_name:
        print("❌ Both fields required!")
        return
    
    session = {
        "id": session_id_counter,
        "vehicle_no": vehicle_no,
        "owner_name": owner_name,
        "entry_time": get_current_time(),
        "exit_time": None,
        "fee": None
    }
    
    sessions.append(session)
    session_id_counter += 1
    
    print(f"✅ Entry recorded: {vehicle_no}")


def checkout():
    """Checkout a vehicle and calculate fee."""
    try:
        sid = int(input("Session ID: "))
    except ValueError:
        print("❌ Invalid ID!")
        return
    
    # Find session
    session = None
    for s in sessions:
        if s["id"] == sid:
            session = s
            break
    
    if not session:
        print("❌ Session not found!")
        return
    
    if session["exit_time"]:
        print("⚠ Already checked out!")
        return
    
    # Set exit time
    session["exit_time"] = get_current_time()
    
    # Calculate fee
    entry = parse_time(session["entry_time"])
    exit_time = parse_time(session["exit_time"])
    hours = (exit_time - entry).seconds / 3600
    hours = max(1, int(hours))  # Minimum 1 hour
    
    session["fee"] = hours * HOURLY_RATE
    
    print(f"🚗 Checkout successful! Fee: ${session['fee']:.2f}")


def search_sessions():
    """Search by vehicle number or owner name."""
    keyword = input("Search keyword: ").strip().lower()
    
    found = False
    for s in sessions:
        if keyword in s["vehicle_no"].lower() or keyword in s["owner_name"].lower():
            print(f"ID: {s['id']} | {s['vehicle_no']} | {s['owner_name']} | Entry: {s['entry_time']}")
            if s["exit_time"]:
                print(f"   Exit: {s['exit_time']} | Fee: ${s['fee']:.2f}")
            else:
                print(f"   Still parked")
            found = True
    
    if not found:
        print("❌ No matching sessions!")


def view_all():
    """View all sessions."""
    if not sessions:
        print("No sessions found.")
        return
    
    print("\n" + "=" * 70)
    print(f"{'ID':<5} {'Vehicle':<12} {'Owner':<12} {'Entry':<20} {'Exit':<20} {'Fee':<10}")
    print("=" * 70)
    
    for s in sessions:
        e_time = s["entry_time"]
        x_time = s["exit_time"] if s["exit_time"] else "Active"
        fee = f"${s['fee']:.2f}" if s["fee"] else "-"
        
        print(f"{s['id']:<5} {s['vehicle_no']:<12} {s['owner_name']:<12} {e_time:<20} {x_time:<20} {fee:<10}")
    
    print("=" * 70)


def billing():
    """Show total revenue."""
    total = 0
    count = 0
    
    for s in sessions:
        if s["fee"]:
            total += s["fee"]
            count += 1
    
    print(f"\n💰 Completed sessions: {count}")
    print(f"💵 Total revenue: ${total:.2f}")


def remove_session():
    """Delete a session."""
    try:
        sid = int(input("Session ID to delete: "))
    except ValueError:
        print("❌ Invalid ID!")
        return
    
    for s in sessions:
        if s["id"] == sid:
            sessions.remove(s)
            print("✅ Session deleted!")
            return
    
    print("❌ Session not found!")


# ======================== Menu ========================
def main():
    while True:
        print("\n" + "=" * 40)
        print("   PARKING LOT SYSTEM")
        print("=" * 40)
        print("1. Add Entry")
        print("2. Checkout")
        print("3. Search")
        print("4. View All")
        print("5. Billing")
        print("6. Remove")
        print("0. Exit")
        print("=" * 40)
        
        choice = input("Choice: ").strip()
        
        if choice == "1":
            add_entry()
        elif choice == "2":
            checkout()
        elif choice == "3":
            search_sessions()
        elif choice == "4":
            view_all()
        elif choice == "5":
            billing()
        elif choice == "6":
            remove_session()
        elif choice == "0":
            print("👋 Goodbye!")
            break
        else:
            print("⚠ Invalid choice!")


if __name__ == "__main__":
    main()
```
