## 1. Develop a simple calculator program that performs basic arithmetic operations (+,
-, *, /) on two numbers provided by the user. The program should ask the user for
the numbers and the operator. However, the program should handle the following
exceptions:
a. Invalid Number: If the user enters a number that is not valid, catch the
exception and display an error message.
b. Invalid Operator: If the user enters an operator other than "+", "-", "*", or
"/", catch the exception and display an error message.
c. Division by Zero: If the user tries to divide by zero, catch the exception and
display an error message.
Write a program that performs the requested arithmetic operation and
handles the exceptions as described above

```python
def calculator():
    try:
        # Get first number
        try:
            num1 = float(input("Enter first number: "))
        except ValueError:
            raise ValueError("Invalid Number: Please enter a valid numeric value.")
        
        # Get operator
        operator = input("Enter operator (+, -, *, /): ").strip()
        if operator not in ['+', '-', '*', '/']:
            raise ValueError("Invalid Operator: Please use +, -, *, or / only.")
        
        # Get second number
        try:
            num2 = float(input("Enter second number: "))
        except ValueError:
            raise ValueError("Invalid Number: Please enter a valid numeric value.")
        
        # Perform calculation
        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            if num2 == 0:
                raise ZeroDivisionError("Division by Zero: Cannot divide by zero.")
            result = num1 / num2
        
        print(f"Result: {num1} {operator} {num2} = {result}")
    
    except ValueError as ve:
        print(f"Error: {ve}")
    except ZeroDivisionError as zde:
        print(f"Error: {zde}")

# Run the calculator
calculator()
```

```python
## 2. Create class television that has members to hold the model number ,screen size
and price. Take a member function to take input from user, If more than 4 digits
are entered for model number, if screen size is smaller than 12 inches or greater
than 70 inches or if the price is negative or greater than 5000 Rs, then throw an
exception.
Write a main() that instantiates an object and allows the user to enter and display
data. If exception is caught, replace all data member values with zero
```

```python
class Television:
    def __init__(self):
        self.model_number = 0
        self.screen_size = 0
        self.price = 0
    
    def input_data(self):
        try:
            # Model number
            model = input("Enter model number (max 4 digits): ")
            if len(model) > 4:
                raise ValueError("Model number cannot have more than 4 digits.")
            if not model.isdigit():
                raise ValueError("Model number must be numeric.")
            self.model_number = int(model)
            
            # Screen size
            size = float(input("Enter screen size (inches): "))
            if size < 12 or size > 70:
                raise ValueError("Screen size must be between 12 and 70 inches.")
            self.screen_size = size
            
            # Price
            price = float(input("Enter price (Rs): "))
            if price < 0 or price > 5000:
                raise ValueError("Price must be between 0 and 5000 Rs.")
            self.price = price
        
        except ValueError as e:
            print(f"Exception caught: {e}")
            print("Resetting all data members to zero.")
            self.model_number = 0
            self.screen_size = 0
            self.price = 0
    
    def display(self):
        print(f"Model Number: {self.model_number}")
        print(f"Screen Size: {self.screen_size} inches")
        print(f"Price: Rs {self.price}")

# Main function
def main():
    tv = Television()
    tv.input_data()
    tv.display()

# Run the program
main()
```
