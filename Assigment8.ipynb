## 1 Write a program to calculate area of rectangle

```python
def area_rectangle(length, breadth):
    return length * breadth

l = float(input("Enter length: "))
b = float(input("Enter breadth: "))

print("Area of Rectangle =", area_rectangle(l, b))
```

## 4. Sum of all odd numbers between 1 to n

```python
def odd_sum(n):
    total = 0

    for i in range(1, n + 1, 2):
        total += i

    return total

n = int(input("Enter n: "))
print("Sum of Odd Numbers =", odd_sum(n))
```

## 5. Sum of all prime numbers between 1 to n

```python
def is_prime(num):
    if num < 2:
        return False

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False

    return True


def prime_sum(n):
    total = 0

    for i in range(2, n + 1):
        if is_prime(i):
            total += i

    return total

n = int(input("Enter n: "))
print("Sum of Prime Numbers =", prime_sum(n))
```

# 6 Write a program to find print the following Fibonacci series using functions:
1 1 2 3 5 8 n terms

```python
def fibonacci(n):
    a = 1
    b = 1

    for i in range(n):
        print(a, end=" ")
        c = a + b
        a = b
        b = c

n = int(input("Enter number of terms: "))
fibonacci(n)
```

## 7. Write a program to find sum of digits of a number.

```python
def sum_digits(num):
    total = 0

    while num > 0:
        total += num % 10
        num //= 10

    return total

n = int(input("Enter number: "))
print("Sum of Digits =", sum_digits(n))
```

## 8. Write a program find reverse of a number

```python
def reverse_number(num):
    rev = 0

    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num //= 10

    return rev

n = int(input("Enter number: "))
print("Reverse =", reverse_number(n))
```

```python
def reverse_number(num):
    rev = 0

    while num > 0:
        rev = rev * 10 + (num % 10)
        num //= 10

    return rev


def is_palindrome(num):
    return num == reverse_number(num)

n = int(input("Enter number: "))

if is_palindrome(n):
    print("Palindrome Number")
else:
    print("Not Palindrome")
```

```python
def is_leap(year):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        return True

    return False

year = int(input("Enter year: "))

if is_leap(year):
    print("Leap Year")
else:
    print("Not a Leap Year")
```
