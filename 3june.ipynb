## demo_exception2

```python
try:
    x = int(input("Enter first number : "))
    y = int(input("Enter first number : "))
    result = x/y    
#Specialized exception handler
except ValueError:
    print("Please enter a number")
except ZeroDivisionError:
    print("Denomenator must not be zero")
#Generalized exception handler
except:
    print("Something went wrong...")
else:
    #The code which must execute when there is not exception
    print("Result ",result)
```

## demo_exception3

```python
try:
    x = int(input("Enter first number : "))
    y = int(input("Enter first number : "))
    try:
        result = x/y    
    except ZeroDivisionError:
        print("Denomenator must not be zero")    
except ValueError:
    print("Please enter a number")
#Generalized exception handler
except:
    print("Something went wrong...")
else:
    #The code which must execute when there is not exception
    print("Result ",result)
```

## demo_exception4

```python
try:
    try:
        x = int(input("Enter first number : "))
        y = int(input("Enter first number : "))        
    except ValueError:
        print("Inner block")
        print("Please enter a number")
        raise #re-throw the exception
except ValueError:
    print("Outer block")
    print("Please enter a number")
```

## demo_finally

```python
try:
    x = int(input("Enter first number : "))
    y = int(input("Enter first number : "))        
    result = x+y
    print("Result = ",result)  
except:
    print("Something went wrong")
#Finally block is executed whether exception occurs or not
finally:
    print("This is a very very very important code")
```

## file_read

```python
try:
    fp = open("myfile.txt","r")
    #It will read all data from file
    #reads 10 bytes
    data = fp.read(10)
    print(data)
except FileNotFoundError:
    print("File does not exist")
else:
    fp.close() #Release the resource
```

## functions

```python
def sum(x,y):
    return x+y

def factorial(x):
    f = 1
    while x>=1:
        f=f*x
        x-=1
    return f

def fibonacii(n):
    a,b= 0,1
    for i in range(n):
        c = a+b
        print(c,end=" ")
        a,b=b,c



print(sum(10, 20))
print(factorial(5))
fibonacii(10)
```
