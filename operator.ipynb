```python
 TIP CALCULATOR
INPUT: 
enter your bill:
how many persons you are 
tip % you want to give:

OUTPUT:
your total bill agter tip:
individual share:
```

```python
bill = float(input("please enter your bill="))
persons = float(input(" how many persons you are ="))
tip = float(input(" tip % you want to give ="))

final_bill = bill+((tip/100)*bill)
share = final_bill/persons

print("your final bill=",final_bill)
print("individaul share=",share)
```
