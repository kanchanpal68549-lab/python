# NESTED IF

```python
age = int(input("Please enter your age: "))

if age > 18:
    print("You can apply for a license.")

    license_type = input("Press 'R' to Renew license and Press 'N' for a New license: ")

    if license_type.upper() == 'R':
        print("You selected Renew license.")
    elif license_type.upper() == 'N':
        print("You selected New license.")
    else:
        print("Invalid selection.")
else:
    print("You cannot apply for a license.")
    print("Good luck next time.")
```
