#1 **Write a program print following patterns**

## (a) Hollow Rectangle

* * * * *
*
*
*
*
* * * * *

```python
rows = 6
cols = 5

for i in range(rows):
    for j in range(cols):
        if i == 0 or i == rows-1 or j == 0 or j == cols-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

## (b) Continuous Number Triangle


```python
num = 1

for i in range(1, 5):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

## (c) Pascal's Triangle

```python
rows = 4

for i in range(rows):
    num = 1

    print(" " * (rows - i), end="")

    for j in range(i + 1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)

    print()
```

## (d) Alphabet Triangle

```python
for i in range(1, 6):
    ch = 65

    for j in range(i):
        print(chr(ch), end=" ")
        ch += 1

    print()
```

## (e) Pyramid of Stars

```python
rows = 5

for i in range(rows):
    print(" " * (rows - i), end="")

    for j in range(2 * i + 1):
        print("*", end=" ")

    print()
```

## (f) Number Pyramid

```python
rows = 5

for i in range(1, rows + 1):
    print(" " * (rows - i), end="")

    for j in range(1, 2 * i):
        print(j, end=" ")

    print()
```

## (g) Alphabet Pyramid

```python
rows = 5

for i in range(1, rows + 1):
    print(" " * (rows - i), end="")

    ch = 65

    for j in range(2 * i - 1):
        print(chr(ch), end=" ")
        ch += 1

    print()
```
