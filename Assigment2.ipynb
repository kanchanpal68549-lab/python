#1 **Convert the time entered in hh,min and sec into seconds.**

```python
hh = int(input("Enter hours: "))
mm = int(input("Enter minutes: "))
ss = int(input("Enter seconds: "))

total_seconds = hh * 3600 + mm * 60 + ss
print("Total seconds =", total_seconds)
```

#2 **Convert temp from Celsius to Fahrenheit. (C/5 = (F-32)/9)**

```python
c = float(input("Enter temperature in Celsius: "))

f = (c * 9/5) + 32
print("Temperature in Fahrenheit =", f)
```

#3 **Convert distant given in feet and inches into meter and centimeter.**

```python
feet = int(input("Enter feet: "))
inches = int(input("Enter inches: "))

total_inches = feet * 12 + inches
cm = total_inches * 2.54
meter = cm / 100

print("Meters =", meter)
print("Centimeters =", cm)
```

#4 **WAP to calculate area of triangle and rectangle**

```python
# Triangle
base = float(input("Enter base of triangle: "))
height = float(input("Enter height of triangle: "))

triangle_area = 0.5 * base * height
print("Area of Triangle =", triangle_area)

# Rectangle
length = float(input("Enter length of rectangle: "))
breadth = float(input("Enter breadth of rectangle: "))

rectangle_area = length * breadth
print("Area of Rectangle =", rectangle_area)
```

#5 **WAP to calculate selling price of book based on cost price and discount.**

```python
cp = float(input("Enter cost price: "))
discount = float(input("Enter discount percentage: "))

discount_amount = cp * discount / 100
sp = cp - discount_amount

print("Selling Price =", sp)
```

#6 **WAP to calculate total salary of employee based on basic, da=10% of basic,
ta=12% of basic, hra=15% of basic.**

```python
basic = float(input("Enter basic salary: "))

da = basic * 10 / 100
ta = basic * 12 / 100
hra = basic * 15 / 100

total_salary = basic + da + ta + hra

print("DA =", da)
print("TA =", ta)
print("HRA =", hra)
print("Total Salary =", total_salary)
```

#7 **Find the sum of three-digit number.**

```python
num = int(input("Enter a three-digit number: "))

digit1 = num // 100
digit2 = (num // 10) % 10
digit3 = num % 10

sum_digits = digit1 + digit2 + digit3

print("Sum of digits =", sum_digits)
```

#8 **Write a program to swap two numbers using third variable.**

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

temp = a
a = b
b = temp

print("After swapping:")
print("a =", a)
print("b =", b)
```

#9 **Write a program to swap two numbers without using third variable.**

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

a = a + b
b = a - b
a = a - b

print("After swapping:")
print("a =", a)
print("b =", b)
```

#10**Write a program to reverse three-digit number.**

```python
num = int(input("Enter a three-digit number: "))

digit1 = num // 100
digit2 = (num // 10) % 10
digit3 = num % 10

reverse = digit3 * 100 + digit2 * 10 + digit1

print("Reversed number =", reverse)
```

#11. **Write a program to accept an integer amount from user and tell minimum
number of notes needed for representing that amount.**

```python
amount = int(input("Enter amount: "))

notes500 = amount // 500
amount %= 500

notes200 = amount // 200
amount %= 200

notes100 = amount // 100
amount %= 100

notes50 = amount // 50
amount %= 50

notes20 = amount // 20
amount %= 20

notes10 = amount // 10
amount %= 10

print("500 Notes =", notes500)
print("200 Notes =", notes200)
print("100 Notes =", notes100)
print("50 Notes =", notes50)
print("20 Notes =", notes20)
print("10 Notes =", notes10)
print("Remaining Amount =", amount)
```
