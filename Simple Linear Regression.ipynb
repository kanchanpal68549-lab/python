```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['YearsExperience'],  df['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
```

```python
x = df['YearsExperience']  # input variable

y = df['Salary']          # output variable
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state= 42, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train.values.reshape(-1, 1), y_train)
```

```python
m = lr.coef_
c = lr.intercept_

print("Slope:", m)
print("Intercept:", c)
```

```python
# y = m*x +c 

y = m * 2.0 + c
print(y)
```

```python
y_pred_train = lr.predict(x_train.values.reshape(-1,1))
```

```python
plt.scatter(x_train,y_train,color="red")
# line chart
plt.scatter(x_train,y_train,color="y")

plt.xlabel('YearsExperience')
plt.ylabel('Salary')
```

```python
plt.scatter(x_train, y_train, color="red")   # Actual data points

plt.plot(
    x_train,
    lr.predict(x_train.values.reshape(-1, 1)),
    color="blue"
)  # Regression line

plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.show()
```

```python

plt.scatter(x_train, y_train, color="red")

plt.plot(x_train, m[0] * x_train + c, color="blue")

plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
#Finding errors
from sklearn.metrics import mean_squared_error,root_mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
print("Mean Squared Error ",mean_squared_error(y_train,y_pred_train))
print("Root Mean Squared Error ",np.sqrt(mean_squared_error(y_train,y_pred_train)))
print("Mean Absolute Error ",mean_absolute_error(y_train,y_pred_train))
print("Mean Absolute Percentage Error ",mean_absolute_percentage_error(y_train,y_pred_train))
```
