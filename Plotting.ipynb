```python
name = ["Payal", "Kanchan", "Kiran", "Junaid", "Virat", "Rahul"]
marks = [90, 91, 45, 55, 36, 10]

def show_value(pct):
    total = sum(marks)
    value = int(round(pct * total / 100.0))
    return f"{value}"

plt.pie(marks, labels=name, autopct=show_value)

plt.title("Student Marks Distribution")
plt.show()
```

```python
data = {"Name":["Kanchan", "Payal", "Rahul", "Virat", "Shubam"],
        "Age":[20,21,18,34,89],
        "sal":[100000,1230000,1212,3222,3334],
        "dept":["HR","IT","Sales","Admin","Dev"],
        "exp":[1,2,3,4,5]
       }
df=pd.DataFrame(data)
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='exp', y='sal', data=df, color='r')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='exp', y='sal', data=df, color='r')

plt.title("Experience vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary")

plt.show()
```

```python
sns.lineplot(x="exp", y="sal", data=df, color="g", marker='d')
plt.show()
```

```python
sns.barplot(x="dept", y="sal", data=df, color="g")

plt.title("Department Wise Salary")
plt.xlabel("Department")
plt.ylabel("Salary")

plt.show()
```

```python
sns.histplot(x="sal", data=df, color="y")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="Age", data=df, color="g")

plt.title("Department Wise Salary")
plt.xlabel("Department")
plt.ylabel("Salary")

plt.show()
```

```python
sns.countplot(x="dept", data=df)
plt.show()
```
