```python
import re

data = '''
101 Janvi Yadav 34000
102 Seema Rajput 25000
103 Ritu Varma 40000
104 Reena Singh 45500                 
105 Rohit Bagel 50000
'''

pattern = re.compile(r"(\d{3})\s([A-Z][a-z]+\s[A-Z][a-z]+)\s(\d+)")

result = re.finditer(pattern, data)

for x in result:
    print(x.group(1))  # Employee ID

    
```

```python
# search will search for first occurance

data = '''
101 Janvi Yadav 34000
102 Seema Rajput 25000
103 Ritu Varma 40000
104 Reena Singh 45500
105 Rohit Bagel 50000
'''
Pattern = re.compile(r"[A-Z][a-z]+")   


result = re.search(Pattern, data)

print(result) 
```

```python
# match will search for  occurance at the beginning of string

data = '''
101 Janvi Yadav 34000
102 Seema Rajput 25000
103 Ritu Varma 40000
104 Reena Singh 45500
105 Rohit Bagel 50000
'''
Pattern = re.compile(r"[A-Z][a-z]+")   


result = re.match(Pattern, data)

print(result) 
```

```python
# match will search for  occurance at the beginning of string

data = '''
101 Janvi Yadav 34000
102 Seema Rajput 25000
103 Ritu Varma 40000
104 Reena Singh 45500
105 Rohit Bagel 50000
'''
Pattern = re.compile(r"[A-Z][a-z]+")   


result = re.sub(Pattern, "name", data)

print(result) 
```

```python
# ^ for looking for pattern towards beginning of string

data = '''
101 Janvi Yadav 34000
102 Seema Rajput 25000
103 Ritu Varma 40000
104 Reena Singh 45500                 
105 Rohit Bagel 50000
'''                                      
# Pattern = re.compile(r"^\d{3}")
Pattern = re.compile(r"^\d{3}", re.MULTILINE)

result = re.findall(Pattern, data)

print(result) 
```

```python
# $ for looking for pattern towards end of string

data = '''
101 Janvi Yadav 34000
102 Seema Rajput 25000
103 Ritu Varma 40000
104 Reena Singh 45500                 
105 Rohit Bagel 50000
'''                                      
Pattern = re.compile(r"\d{5}$")

result = re.findall(Pattern, data)

print(result) 
```
