```python
class Time:
    def setTime(self):
        self.hr = int(input("Enter hr: "))
        self.min = int(input("Enter min: "))
        self.sec = int(input("Enter sec: "))

    def showTime(self):
        print(f"{self.hr:02d}:{self.min:02d}:{self.sec:02d}")


t1 = Time()
t1.setTime()
t1.showTime()
```

```python
class Student:
    def __init__(self, sid=1, sname="", percentage=1):
        self.sid = sid
        self.sname = sname
        self.percentage = percentage

    def showStudent(self):
        print("Student ID:", self.sid)
        print("Student Name:", self.sname)
        print("Percentage:", self.percentage)
        print("-" * 30)


s1 = Student()
s1.showStudent()

s2 = Student(301, "Ajay", 76)
s2.showStudent()
```

```python
class Book:
    # Constructor
    def __init__(self, bid=0, bname="", price=0.0, author=""):
        self.bid = bid
        self.bname = bname
        self.price = price
        self.author = author
        print("Book Object Created")

    # Display Method
    def ShowBook(self):
        print("\nBook Details")
        print("Book ID :", self.bid)
        print("Book Name :", self.bname)
        print("Price :", self.price)
        print("Author :", self.author)

    # Destructor
    def __del__(self):
        print("Book Object Destroyed")


# Parameterless Constructor
b1 = Book()
b1.ShowBook()

# Parameterized Constructor
b2 = Book(101, "Python Programming", 450, "Guido van Rossum")
b2.ShowBook()
```

```python
class Product:
    # Constructor
    def __init__(self, pid=0, pname="", price=0.0, quantity=0):
        self.pid = pid
        self.pname = pname
        self.price = price
        self.quantity = quantity
        print("Product Object Created")

    # Display Method
    def ShowProduct(self):
        print("\nProduct Details")
        print("Product ID :", self.pid)
        print("Product Name :", self.pname)
        print("Price :", self.price)
        print("Quantity :", self.quantity)

    # Destructor
    def __del__(self):
        print("Product Object Destroyed")


# Parameterless Constructor
p1 = Product()
p1.ShowProduct()

# Parameterized Constructor
p2 = Product(201, "Laptop", 55000, 5)
p2.ShowProduct()
```

```python
import math
from abc import ABC, abstractmethod

# Base class (Abstract)
class Shape(ABC):
    @abstractmethod
    def calculateArea(self):
        pass

# Subclass for Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculateArea(self):
        return self.width * self.height

# Subclass for Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculateArea(self):
        return math.pi * (self.radius ** 2)

# Subclass for Triangle
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def calculateArea(self):
        return 0.5 * self.base * self.height

# Example usage:
shapes = [
    Rectangle(10, 5),
    Circle(7),
    Triangle(10, 8)
]

for shape in shapes:
    print(f"{type(shape).__name__}: {shape.calculateArea()}")
```
