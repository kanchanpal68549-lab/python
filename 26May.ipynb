```python
class Student:
    def setStudent(self):
        self.sid = int(input("Enter sid : "))
        self.sname = input("Enter sname : ")
        self.percentage = float(input("Enter percentage : "))

    def showStudent(self):
        print("Sid = ",self.sid)
        print("SName = ",self.sname)
        print("Percentage =",self.percentage)

s1 = Student()
s1.setStudent()
s1.showStudent()
```

```python
class Time:
    def setTime(self):
        self.hr = int(input("Enter hr : "))
        self.min = int(input("Enter min : "))
        self.sec = int(input("Enter sec : "))

    def showTime(self):
        print(self.hr," : ",self.min," : ",self.sec)

t1 = Time()
t1.setTime()
t1.showTime()
```
