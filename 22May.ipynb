## armstong

```python
num = int(input("Enter any number : "))
count = 0
temp = num
while num!= 0:
    num = num // 10
    count += 1

print("Number of digits =",count)

sum = 0
num = temp
while num != 0:
    sum = sum + (num%10)**count
    num = num // 10

if sum == temp:
    print(temp," is armstrong")
else:
    print(temp," is not armstrong")
```
