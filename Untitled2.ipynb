```python
class Emp:
    def __init__(self, id, nm, bs):
        self.eid = id
        self.ename = nm
        self.basic = bs

    def display(self):
        print("Eid =", self.eid)
        print("EName =", self.ename)
        print("Basic =", self.basic)


class Company:
    def __init__(self):
        self.allEmps = {}

    def addEmp(self, e):
        self.allEmps[e.eid] = e

    def display(self):
        for emp in self.allEmps.values():
            emp.display()
            print("--------------------")

    def removeEmp(self, eid):
        if eid in self.allEmps:
            del self.allEmps[eid]
            print("Employee deleted")
        else:
            print("Employee not found")

    def editEmp(self, eid):
        if eid in self.allEmps:
            ename = input("Enter new name: ")
            basic = float(input("Enter new salary: "))
            self.allEmps[eid].ename = ename
            self.allEmps[eid].basic = basic
            print("Employee updated")
        else:
            print("Employee not found")

    def save(self):
        print("Data saved successfully")


choice = 0
cmp = Company()

while choice != 6:
    print("\n\t\t1. Add new emp")
    print("\t\t2. Display emps")
    print("\t\t3. Remove emp")
    print("\t\t4. Edit emp")
    print("\t\t6. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        eid = int(input("Enter eid: "))
        ename = input("Enter ename: ")
        basic = float(input("Enter basic: "))
        e1 = Emp(eid, ename, basic)
        cmp.addEmp(e1)

    elif choice == 2:
        cmp.display()

    elif choice == 3:
        eid = int(input("Enter eid to delete: "))
        cmp.removeEmp(eid)

    elif choice == 4:
        eid = int(input("Enter eid to edit: "))
        cmp.editEmp(eid)

    elif choice == 6:
        cmp.save()
        print("-------End of program------")

    else:
        print("Invalid choice")


choice = 0

while choice != 3:
    print("\n\t\t1. Add new content")
    print("\t\t2. Read content")
    print("\t\t3. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        fp = open("myfile.txt", "a")
        data = input("Enter content: ")
        fp.write(data + "\n")
        fp.close()

    elif choice == 2:
        try:
            fp = open("myfile.txt", "r")
            print(fp.read())
            fp.close()
        except FileNotFoundError:
            print("File does not exist")

    elif choice == 3:
        print("------------ End of program ------------")

    else:
        print("Invalid choice")
```

```python
import pickle

class Emp:
    def __init__(self, eno, ename, salary):
        self.eno = eno
        self.ename = ename
        self.salary = salary

    def display(self):
        print("Emp No:", self.eno)
        print("Name:", self.ename)
        print("Salary:", self.salary)

with open("empData.dat", "rb") as fp:
    e1 = pickle.load(fp)

e1.display()
```
