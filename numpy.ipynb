```python
#  WHAT IS NUMPY?
Numpy is a Python external library that helps us perform various scientific Computations and allows us to work
with multidimensional arrays more efficiently.
After learning that Numpy is used to working with multidimensional arrays you might wonder
why we use this external library if we have a Python list. We can also use List to work with multidimensional arrays.
So, this is a very valid question. I will tell you why we use numpy if we have a Python list. let’s explore the difference.
```

```python

import numpy as np

a = np.array(42)
b = np.array([1, 2, 3, 4, 5]) 
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])
print(a.ndim)
print(b.ndim) 
print(c.ndim) 
print(d.ndim)
```
