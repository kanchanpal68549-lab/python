```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(1,11)
y = x**2
print(x)
print(y)
```

```python
# Line plot
plt.plot(x,y,color='g',linestyle='dashdot')
```

```python
# Bar plot
plt.bar(x,y)
```

```python
plt.bar(x,y,color=['pink','skyblue'] )
```

```python
#Horizontal plot
plt.barh(x,y,color=['pink','skyblue'] )
```

```python
plt.scatter(x,y)
```

```python

# * , p , h, o
plt.scatter(x,y,color='g',marker="^")
```

```python
plt.plot(x, y, color='g', marker='*', linestyle='--')
```

```python
plt.scatter(x, y, color='g', marker='o', s=200)
```

```python
plt.plot(x, y, color='black')
plt.bar(x,y,color=['pink','yellow'])
plt.scatter(x,y,color='skyblue',marker='^')
```

```python
plt.plot(x, y, color='black', label='Sales')
plt.bar(x,y,color=['pink','yellow'])
plt.scatter(x,y,color='skyblue',marker='^')
plt.title("This is my sample graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
```

```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=0
)

plt.title("Programming Language Popularity")
plt.show()
```
