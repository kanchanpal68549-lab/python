```python
ccnum = '1234-5678-5678'
output:
  12xx-xxxx-xx78
```

```python
ccnum = '1234-5678-5678'

# Split into parts by '-'
parts = ccnum.split('-')

# Masking the digits
masked = (
    parts[0][:2] + 'xx' + '-' +   # Keep first 2 digits, mask next 2
    'xxxx' + '-' +                # Fully mask second group
    'xx' + parts[2][2:]           # Mask first 2 digits, keep last 2
)

print(masked)
```
