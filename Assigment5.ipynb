#1. **1. Write a program to prompt user to enter userid and password. If Id and
password is incorrect give him chance to re-enter the credentials. Let him try 3
times. After that program to terminate.**

```python
userid = "admin"
password = "1234"

for i in range(3):
    uid = input("Enter User ID: ")
    pwd = input("Enter Password: ")

    if uid == userid and pwd == password:
        print("Login Successful")
        break
    else:
        print("Invalid Credentials")

else:
    print("Maximum attempts reached. Program terminated.")
```

```python
# Program 1: User login with 3 attempts
correct_userid = "admin"
correct_password = "password123"

attempts = 0
max_attempts = 3

while attempts < max_attempts:
    userid = input("Enter userid: ")
    password = input("Enter password: ")
    if userid == correct_userid and password == correct_password:
        print("Login successful!")
        break
    else:
        attempts += 1
        remaining = max_attempts - attempts
        if remaining > 0:
            print(f"Incorrect credentials. {remaining} attempt(s) left.")
        else:
            print("Too many failed attempts. Program terminated.")
```

#2 **Enter number of students from user. For those many students accept marks of 5subject marks from user and calculate percentage. Display all percentage andaverage percentage of students.**

```python
n = int(input("Enter number of students: "))

total_percentage = 0

for i in range(n):
    total_marks = 0

    print(f"\nStudent {i+1}")

    for j in range(5):
        marks = float(input(f"Enter marks of subject {j+1}: "))
        total_marks += marks

    percentage = total_marks / 5
    total_percentage += percentage

    print("Percentage =", percentage)

avg_percentage = total_percentage / n

print("\nAverage Percentage of Students =", avg_percentage)
```

```python
# Program 2: Student marks, percentages, and class average
num_students = int(input("Enter number of students: "))
all_percentages = []

for i in range(num_students):
    print(f"\n--- Student {i+1} ---")
    marks = []
    for j in range(5):
        mark = float(input(f"Enter marks for subject {j+1}: "))
        marks.append(mark)
    percentage = sum(marks) / 5   # assuming total marks per subject = 100
    all_percentages.append(percentage)
    print(f"Percentage: {percentage:.2f}%")

average_percentage = sum(all_percentages) / num_students
print("\nAll percentages:")
for idx, perc in enumerate(all_percentages):
    print(f"Student {idx+1}: {perc:.2f}%")
print(f"Average percentage of all students: {average_percentage:.2f}%")
```

#3 **Accept no. of passengers from user and per ticket cost. Then accept age of eachpassenger and then calculate total amount to ticket to travel for all of them based onfollowing condition :**
a. Children below 12 = 30% discount
b. Senior citizen (above 59) = 50% discount
c. Others need to pay full.

```python
# Program 3: Passenger ticket cost with discounts
num_passengers = int(input("Enter number of passengers: "))
ticket_cost = float(input("Enter per ticket cost: "))

total_amount = 0
for i in range(num_passengers):
    age = int(input(f"Enter age of passenger {i+1}: "))
    if age < 12:
        discount = 0.30
    elif age > 59:
        discount = 0.50
    else:
        discount = 0.00
    fare = ticket_cost * (1 - discount)
    total_amount += fare
    print(f"Passenger {i+1} pays: {fare:.2f}")

print(f"\nTotal amount to collect: {total_amount:.2f}")
```

```python
passengers = int(input("Enter number of passengers: "))
ticket_cost = float(input("Enter ticket cost: "))

total_amount = 0

for i in range(passengers):
    age = int(input(f"Enter age of passenger {i+1}: "))

    if age < 12:
        fare = ticket_cost * 0.70
    elif age > 59:
        fare = ticket_cost * 0.50
    else:
        fare = ticket_cost

    total_amount += fare

print("Total Amount =", total_amount)
```

#4 **WAP to print Armstrong number within a given range**

```python
start = int(input("Enter start range: "))
end = int(input("Enter end range: "))

for num in range(start, end + 1):
    temp = num
    digits = len(str(num))
    arm_sum = 0

    while temp > 0:
        digit = temp % 10
        arm_sum += digit ** digits
        temp //= 10

    if arm_sum == num:
        print(num)
```

```python
# Program 4: Armstrong numbers in a range
def is_armstrong(num):
    digits = str(num)
    n = len(digits)
    total = sum(int(d) ** n for d in digits)
    return total == num

start = int(input("Enter start of range: "))
end = int(input("Enter end of range: "))
print(f"Armstrong numbers between {start} and {end}:")
for num in range(start, end + 1):
    if is_armstrong(num):
        print(num)
```

#5 **Write a program to print prime numbers between 1 to 100.**

```python
for num in range(2, 101):
    prime = True

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            prime = False
            break

    if prime:
        print(num, end=" ")
```

```python
# Program 5: Prime numbers from 1 to 100
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print("Prime numbers between 1 and 100:")
for num in range(1, 101):
    if is_prime(num):
        print(num, end=" ")
print()
```

#6 **Write a program to print first n prime numbers.**


```python
n = int(input("Enter n: "))

count = 0
num = 2

while count < n:
    prime = True

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            prime = False
            break

    if prime:
        print(num, end=" ")
        count += 1

    num += 1
```

```python
# Program 6: First n prime numbers
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

n = int(input("Enter n: "))
primes = []
num = 2
while len(primes) < n:
    if is_prime(num):
        primes.append(num)
    num += 1
print(f"First {n} prime numbers: {primes}")
```

#7 **Write a program to solve the following series**
a. 1! + 2! + 3! + 4! + .....n!
b. N + N^2 + N^3+N^4 .....+N^N (here ^ means exponent)
c. Find the sum of a geometric series from 1 to n where the common ratio is 2.
d. S = a + a2 / 2 + a3 / 3 + ...... + a10 / 10
e. x - x2/3 + x3/5 - x4/7 + .... to n terms

```python
# 7a: Sum of factorials
import math
n = int(input("Enter n: "))
total = sum(math.factorial(i) for i in range(1, n+1))
print(f"Sum of factorials 1! to {n}! = {total}")
```

```python
# 7b: Sum of powers
N = float(input("Enter N: "))
total = sum(N**i for i in range(1, int(N)+1))
print(f"Sum = {total}")
```

```python
# 7c: Geometric series: 1 + 2 + 4 + ... + 2^(n-1)
n = int(input("Enter n (number of terms): "))
total = sum(2**i for i in range(n))   # 2^0 + 2^1 + ... + 2^(n-1)
print(f"Sum of geometric series (ratio 2, {n} terms) = {total}")
```

```python
# 7d: Series with division
a = float(input("Enter a: "))
total = sum((a**i) / i for i in range(2, 11)) + a   # because a^1/1 = a
print(f"S = {total}")
```

```python
# 7e: Alternating series with odd denominators
x = float(input("Enter x: "))
n = int(input("Enter number of terms: "))
total = 0
for i in range(1, n+1):
    term = (x**i) / (2*i - 1)
    if i % 2 == 0:
        total -= term
    else:
        total += term
print(f"Sum = {total}")
```

```python
#7(E)

x = float(input("Enter x: "))
n = int(input("Enter number of terms: "))

s = 0
sign = 1

for i in range(1, n + 1):
    denominator = 2 * i - 1

    term = sign * (x ** i) / denominator
    s += term

    sign *= -1

print("Sum =", s)
```
