 ## Customize Chart

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [20, 25, 30, 55, 40]

plt.figure(figsize=(4,3))  #figure size 

plt.plot(x, y, 
         color='green',       # line color
         marker='^',         # marker style (not marks!)
         linestyle='dashed', # line style
         linewidth=2,       # line width
         markersize=10)      # marker size


plt.show()
```

```python
plt.plot(x, y, 
         color='#008',      # line color
         marker='d',         # marker style (not marks!)
         linestyle='-.', # line style
         linewidth=2,       # line width
         markersize=10)      # marker size



plt.show()
```

## 3 Advance line chart - multiple Lines & Legends

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [20, 25, 30, 55, 40]
y2 = [30, 35, 40, 65, 55]

plt.figure(figsize=(6,4))

plt.plot(x, y1, label="Sales 2024", marker='o')
plt.plot(x, y2, label="Sales 2026", marker='s')

plt.title("Hotel Sales")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.legend()
plt.grid(True) 

plt.show()
```

## Histogram

```python
 # used for distribution analysis
import matplotlib.pyplot as plt
import random

# Generate 100 random numbers between 1 and 10
data = [random.randint(1, 10) for i in range(100)]

plt.hist(data, bins=10)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

## Pie Chart

```python
# Used to show part to whole relationships/data

Catagories = ['A','B','C','D','E']
sales = [10,20, 55, 35, 45]

plt.pie(sales, labels=Catagories, autopct='%1.1f%%')

plt.title('Pie Chart')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x, y)

z = np.sin(x**2 + y**2)

fig = plt.figure()

ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(x, y, z, cmap="viridis")

ax.set_axis_off()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# surface plot with color representing z-values (4th dimension)
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add colorbar to show the value scale
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='Z value')

ax.axis('off')
plt.show()
```

```python
import matplotlib.animation as animation

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

def update(frame):
    ax.clear()
    ax.axis('off')
    # Rotate the view
    ax.view_init(elev=30, azim=frame)
    ax.plot_surface(X, Y, Z, cmap='viridis')
    return []

ani = animation.FuncAnimation(fig, update, frames=range(0, 360, 5), interval=50)
plt.show()
```
