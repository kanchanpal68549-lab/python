## demo_exception1

```python
try:
    x = int(input("Enter a number : "))
    print("You entered :", x)
except:
    print("Something went wrong...")
```

## demo_exception2

```python
try:
    x = int(input("Enter first number : "))
    y = int(input("Enter first number : "))
    result = x/y    
#Specialized exception handler
except ValueError:
    print("Please enter a number")
except ZeroDivisionError:
    print("Denomenator must not be zero")
#Generalized exception handler
except:
    print("Something went wrong...")
else:
    #The code which must execute when there is not exception
    print("Result ",result)
```

## demo_exception3

```python
try:
    x = int(input("Enter first number : "))
    y = int(input("Enter first number : "))
    try:
        result = x/y    
    except ZeroDivisionError:
        print("Denomenator must not be zero")    
except ValueError:
    print("Please enter a number")
#Generalized exception handler
except:
    print("Something went wrong...")
else:
    #The code which must execute when there is not exception
    print("Result ",result)
```

## demo_exception4

```python
try:
    try:
        x = int(input("Enter first number : "))
        y = int(input("Enter first number : "))        
    except ValueError:
        print("Inner block")
        print("Please enter a number")
        raise #re-throw the exception
except ValueError:
    print("Outer block")
    print("Please enter a number")
```

## demo_finally

```python
try:
    x = int(input("Enter first number : "))
    y = int(input("Enter first number : "))        
    result = x+y
    print("Result = ",result)  
except:
    print("Something went wrong")
#Finally block is executed whether exception occurs or not
finally:
    print("This is a very very very important code")
```

## demo_module1

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

print("Factorial =", factorial(4))
```

## demo_module3

```python
def fibonacii(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
```
