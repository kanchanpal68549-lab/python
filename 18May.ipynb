# Calculator.py

```python
num1 = int(input("Enter a number :"))
num2 = int(input("Enter a number : "))
op = input("Enter operator (+,-,/,*) : ")

if op == "+":
    print("Sum = ",num1+num2)
elif op == '-':
    print("Difference =",num1-num2)
elif op == '*':
    print("Product :",num1*num2)
elif op == '/':
    print("Division :",num1/num2)
else:
    print("Invalid operator")
```
