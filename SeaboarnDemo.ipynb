```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
iris = pd.read_csv('Iris.csv')
```

```python
iris.head()
```

```python
iris.Species.value_counts()
```

```python
sns.countplot(x=iris['Species'],hue=iris['Species'])
```

```python
sns.lineplot(data=iris,x='SepalLengthCm',y='SepalWidthCm')
```

```python
sns.lineplot(data=iris,x='PetalLengthCm',y='PetalWidthCm',hue='Species')
```

```python
sns.scatterplot(data=iris,x='PetalLengthCm',y='PetalWidthCm',hue='Species')
```

```python
sns.scatterplot(data=iris,x='SepalLengthCm',y='SepalWidthCm',hue='Species')
```

```python
sns.histplot(data=iris,x='SepalLengthCm',hue='Species')
```

```python
sns.histplot(data=iris,x='PetalLengthCm',hue='Species')
```

```python
sns.set_style("whitegrid")
sns.histplot(data=iris,x='PetalLengthCm',hue='Species',kde=True)
```

```python
sns.barplot(data=iris,x='Species',y='PetalLengthCm',hue='Species')
```

```python
sns.barplot(
    data=iris,
    y='Species',
    x='PetalLengthCm',
    hue='Species'
)
```

```python
sns.boxplot(data=iris,x='Species',y='PetalLengthCm',hue='Species')
```

```python
#It is a combination of histogram and box plot
sns.violinplot(data=iris,x='Species',y='PetalLengthCm',hue='Species')
```
