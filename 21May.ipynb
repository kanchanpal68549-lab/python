```python
# Number Pyramid
```

```python
n = 5

for i in range(1, n + 1):

    # Spaces
    for j in range(n - i):
        print(" ", end="")

    # Increasing numbers
    for j in range(1, i + 1):
        print(j, end="")

    # Decreasing numbers
    for j in range(i - 1, 0, -1):
        print(j, end="")

    print()
```

```python
## Floyd's Tringle
```

```python
n = 5
num = 1

for i in range(1, n + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
## Star Pyramid
```

```python
n = 5

for i in range(1, n + 1):
    # Spaces
    print(" " * (n - i), end="")

    # Stars
    print("*" * (2 * i - 1))
    
```

```python
## Hellow Pyramid
```

```python
n = 5

for i in range(1, n + 1):

    # Spaces
    for j in range(n - i):
        print(" ", end="")

    # Stars and hollow spaces
    for j in range(1, 2 * i):
        if j == 1 or j == 2 * i - 1 or i == n:
            print("*", end="")
        else:
            print(" ", end="")

    print()
```

```python
## Butterfly Pattern
```

```python
n = 4

# Upper half
for i in range(1, n + 1):
    print("*" * i + " " * (2 * (n - i)) + "*" * i)

# Lower half
for i in range(n - 1, 0, -1):
    print("*" * i + " " * (2 * (n - i)) + "*" * i)
```

```python
## Pascal's Triangle
```

```python
n = 5

for i in range(n):
    num = 1

    # Spaces
    print(" " * (n - i), end="")

    for j in range(i + 1):
        print(num, end="  ")

        num = num * (i - j) // (j + 1)

    print()
```

```python
## Palindrome Number Pyramid
```

```python
n = 5

for i in range(1, n + 1):

    # Spaces
    for j in range(n - i):
        print(" ", end="")

    # Increasing numbers
    for j in range(1, i + 1):
        print(j, end="")

    # Decreasing numbers
    for j in range(i - 1, 0, -1):
        print(j, end="")

    print()
```
