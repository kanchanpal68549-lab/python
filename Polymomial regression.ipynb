```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Position_Salaries.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Level'],df['Salary'])
plt.title('Level vs Salary')
plt.xlabel('Lavel')
plt.ylabel('Salary')
```

```python
x = df['Level']
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
```

```python
regressor.fit(x.values.reshape(-1,1),y)
```

```python
y_pred = regressor.predict(x.values.reshape(-1,1))
```

```python
plt.plot(x.values.reshape(-1,1), y_pred, color='red')
plt.scatter(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=6)

poly_reg.fit(x.values.reshape(-1, 1))
x_poly = poly_reg.transform(x.values.reshape(-1, 1))
```

```python
regressor2 = LinearRegression()
regressor2.fit(x_poly,y)
```

```python
y_pred2 = regressor2.predict(x_poly)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred2,color='red')
plt.title('Level vs Salary')
plt.xlabel('Lavel')
plt.ylabel('Salary')
```
