#4. **WAP to print factorial of a number**

```python
n = int(input("Enter a number: "))

fact = 1
for i in range(1, n + 1):
    fact *= i

print("Factorial =", fact)
```

#4 **WAP to print Fibonacci series upto n.**

```python
n = int(input("Enter number of terms: "))

a = 0
b = 1

for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```

#5 **WAP to print Fibonacci series upto n.**

```python

n = int(input("Enter a number: "))

count = 0

for i in range(1, n + 1):
    if n % i == 0:
        count += 1

if count == 2:
    print("Prime Number")
else:
    print("Not Prime Number")
```

## Eficient method

```python
n = int(input("Enter a number: "))

flag = True

if n < 2:
    flag = False
else:
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            flag = False
            break

if flag:
    print("Prime")
else:
    print("Not Prime")
```

#10 **WAP to check if given number is Perfect Number.**

```python
n = int(input("Enter a number: "))

sum_div = 0

for i in range(1, n):
    if n % i == 0:
        sum_div += i

if sum_div == n:
    print("Perfect Number")
else:
    print("Not Perfect Number")
```

#11 **WAP to check if given number Strong Number.**

```python
n = int(input("Enter a number: "))

temp = n
sum_fact = 0

while temp > 0:
    digit = temp % 10

    fact = 1
    for i in range(1, digit + 1):
        fact *= i

    sum_fact += fact
    temp //= 10

if sum_fact == n:
    print("Strong Number")
else:
    print("Not Strong Number")
```
