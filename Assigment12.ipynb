## 1. Python Program to Replace all Occurrences of ‘a’ with $ in a String

```python
s = input("Enter a string: ")

result = s.replace('a', '$')

print("Modified String:", result)
```

## 2. Python Program to Remove the nth Index Character from a Non-Empty String

```python
s = input("Enter a string: ")
n = int(input("Enter index to remove: "))

result = s[:n] + s[n+1:]

print("Result:", result)
```

## 3. Python Program to Detect if Two Strings are Anagrams

```python
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if sorted(s1.lower()) == sorted(s2.lower()):
    print("Anagram")
else:
    print("Not Anagram")
```

## 4. Python Program to Form a New String where the First Character and the Last Character have been Exchanged

```python
s = input("Enter a string: ")

if len(s) > 1:
    result = s[-1] + s[1:-1] + s[0]
else:
    result = s

print("Result:", result)
```

## 5. Python Program to Count the Number of Vowels in a String

```python
s = input("Enter a string: ")

count = 0

for ch in s.lower():
    if ch in "aeiou":
        count += 1

print("Number of vowels =", count)
```

## 6. Python Program to Take in a String and Replace Every Blank Space with Hyphen

## 7. Python Program to Calculate the Length of a String Without Using a Library Function

```python
s = input("Enter a string: ")

count = 0

for ch in s:
    count += 1

print("Length =", count)
```

## 8. Python Program to Remove the Characters of Odd Index Values in a String

```python
s = input("Enter a string: ")

result = ""

for i in range(len(s)):
    if i % 2 == 0:
        result += s[i]

print(result)
```

## 9. Python Program to Calculate the Number of Words and the Number of Characters Present in a String

## 10.Python Program to Take in Two Strings and Display the Larger String without Using Built-in Functions

```python
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

count1 = 0
for ch in s1:
    count1 += 1

count2 = 0
for ch in s2:
    count2 += 1

if count1 > count2:
    print("Larger String =", s1)
elif count2 > count1:
    print("Larger String =", s2)
else:
    print("Both strings have same length")
```

## 11. Python Program to replace every blank space with hyphen in a string.

```python
s = input("Enter a string: ")

new_str = ""

for ch in s:
    if ch == " ":
        new_str += "-"
    else:
        new_str += ch

print(new_str)
```

## 12. Python Program to count number of lowercase characters in a string.

```python
s = input("Enter a string: ")

count = 0

for ch in s:
    if 'a' <= ch <= 'z':
        count += 1

print("Lowercase characters =", count)
```

## 13. Python Program to count number of digits and letters in a string.

```python
s = input("Enter a string: ")

digits = 0
letters = 0

for ch in s:
    if ch.isdigit():
        digits += 1
    elif ch.isalpha():
        letters += 1

print("Letters =", letters)
print("Digits =", digits)
```

## 14. Python Program to count the occurrences of ach word in a string.

```python
s = input("Enter a string: ")

words = s.split()

for word in set(words):
    print(word, "=", words.count(word))
```

## 15. Python Program to find larger string without using built-in functions.

```python
def string_length(s):
    count = 0

    for ch in s:
        count += 1

    return count

s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if string_length(s1) > string_length(s2):
    print("Larger String:", s1)
elif string_length(s2) > string_length(s1):
    print("Larger String:", s2)
else:
    print("Both strings are of equal length")
```
