## 1. Write a program to find sum of following series using recursive functions:

i. 1! + 2! + 3! + 4! +..... + n!
Note : For fact and sum two recursive functions

```python
# Recursive factorial
def fact(n):
    if n == 0 or n == 1:
        return 1
    return n * fact(n - 1)

# Recursive sum of factorials from 1 to n
def sum_fact_series(n):
    if n == 1:
        return fact(1)
    return fact(n) + sum_fact_series(n - 1)

# Demo
n = int(input("Enter n: "))
print(f"Sum of factorials from 1! to {n}! = {sum_fact_series(n)}")
```

## 2. Write a program to check if given number is Armstrong or not using recursive function.

```python
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n // 10)

def armstrong_sum(n, power):
    if n == 0:
        return 0
    return (n % 10) ** power + armstrong_sum(n // 10, power)

def is_armstrong(num):
    if num < 0:
        return False
    power = count_digits(num)
    return armstrong_sum(num, power) == num

# Demo
num = int(input("Enter a number: "))
if is_armstrong(num):
    print(f"{num} is an Armstrong number.")
else:
    print(f"{num} is not an Armstrong number.")
```

## 3. Write a program to reverse a given number using recursive function.


```python
def reverse_num(n, rev=0):
    if n == 0:
        return rev
    return reverse_num(n // 10, rev * 10 + n % 10)

# Demo
num = int(input("Enter a number: "))
print(f"Reversed number = {reverse_num(abs(num))}{'-' if num < 0 else ''}")
```

## 4. Write a program to find sum of n numbers using recursion.

```python
def sum_n(n):
    if n == 0:
        return 0
    return n + sum_n(n - 1)

# Demo
n = int(input("Enter n: "))
print(f"Sum of first {n} natural numbers = {sum_n(n)}")
```

## 5. Write a program to find factorial using recursion.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

# Demo
n = int(input("Enter n: "))
print(f"{n}! = {factorial(n)}")
```

## 6. Write a program to print Fibonacci series using recursion.

```python
def fib_term(n):
    if n <= 1:
        return n
    return fib_term(n - 1) + fib_term(n - 2)

def print_fib_series(n):
    for i in range(n):
        print(fib_term(i), end=" ")
    print()

# Demo
n = int(input("Enter number of terms: "))
print("Fibonacci series:")
print_fib_series(n)
```

## 7. Write a program to find sum of digits using recursion.

```python
def sum_of_digits(n):
    n = abs(n)
    if n == 0:
        return 0
    return (n % 10) + sum_of_digits(n // 10)

# Demo
num = int(input("Enter a number: "))
print(f"Sum of digits = {sum_of_digits(num)}")
```

## 8. Write a program to check whether a number is prime or not using recursion.

```python
def is_prime_recursive(n, divisor=None):
    if n < 2:
        return False
    if divisor is None:
        divisor = int(n ** 0.5)
    if divisor == 1:
        return True
    if n % divisor == 0:
        return False
    return is_prime_recursive(n, divisor - 1)

# Demo
num = int(input("Enter a number: "))
if is_prime_recursive(num):
    print(f"{num} is prime.")
else:
    print(f"{num} is not prime.")
```

## 9. Write a program to calculate the m to the power n using recursion.

```python
def power(m, n):
    if n == 0:
        return 1
    return m * power(m, n - 1)

# Demo
m = int(input("Enter base (m): "))
n = int(input("Enter exponent (n): "))
print(f"{m}^{n} = {power(m, n)}")
```

## 10. Write a program to reverse a number using recursion.

```python
def reverse_string_recursive(num_str):
    if len(num_str) == 0:
        return ""
    return num_str[-1] + reverse_string_recursive(num_str[:-1])

def reverse_number_alt(num):
    sign = -1 if num < 0 else 1
    rev_str = reverse_string_recursive(str(abs(num)))
    return sign * int(rev_str)

# Demo
num = int(input("Enter a number: "))
print(f"Reversed number = {reverse_number_alt(num)}")
```
